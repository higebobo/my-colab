## Import section

```python
import pandas as pd
```

## Get symbol list

### JPX (https://www.jpx.co.jp/)

```python
url = 'https://www.jpx.co.jp/markets/statistics-equities/misc/tvdivq0000001vg2-att/data_j.xls'
df_jpx = pd.read_excel(url)
df_jpx = df_jpx.iloc[:, [1, 2, 3, 9]]

df_jpx
```

#### Get TSE Prime symbol list

```python
df_tse_prim = df_jpx.loc[df_jpx["市場・商品区分"].str.contains('プライム')]
df_tse_prim = df_tse_prim.iloc[:, [0, 1]]
df_tse_prim.columns = ['symbol', 'name']

df_tse_prim
```

#### Get TSE Standard symbol list

```python
df_tse_std = df_jpx.loc[df_jpx["市場・商品区分"].str.contains('スタンダード')]
df_tse_std = df_tse_std.iloc[:, [0, 1]]
df_tse_std.columns = ['symbol', 'name']

df_tse_std
```

#### Get Japanese company stock symbol list

```python
from investpy.stocks import get_stocks

df_investpy_jp = get_stocks(country='Japan')
df_investpy_jp = df_investpy_jp.iloc[:, [5, 2]]

df_investpy_jp

```

## Merge Japanese and English name

```python
df_tse_prim['symbol'] = df_tse_prim['symbol'].astype(str)
df_investpy_jp['symbol'] = df_investpy_jp['symbol'].astype(str)
df_prim = pd.merge(df_tse_prim, df_investpy_jp, on='symbol').rename(columns={'full_name': 'name(en)'}).set_index('symbol')

df_prim
```
