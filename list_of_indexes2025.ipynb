## Import section

```python
from io import StringIO
from typing import Any
import pandas as pd
import requests
```

## 東証

### 日経225

```python
url = "https://site1.sbisec.co.jp/ETGate/?OutSide=on&_ControlID=WPLETmgR001Control&_PageID=WPLETmgR001Mdtl20&_DataStoreID=DSWPLETmgR001Control&_ActionID=DefaultAID&getFlg=on&burl=search_market&cat1=market&cat2=none&dir=info&file=market_meigara_225.html"
table_list: list = pd.read_html(url, encoding="cp932")
df: pd.DataFrame = table_list[1].iloc[:, [0, 1]]
df = df.drop(0)

df
```

### JPX150

```python
url = "https://www.jpx.co.jp/automation/markets/indices/jpx-prime150/files/jpxprime150weight_j.csv"
df = pd.read_csv(url, encoding="cp932")
df = df.iloc[1:-11, [2, 1]]
df[df.columns[0]] = df[df.columns[0]].astype(str).str[:4]

df
```

### JPX400

```python
url = "https://indexes.nikkei.co.jp/nkave/archives/file/jpx_nikkei_index_400_weight_jp.csv"
response: Any = requests.get(url)
decoded_content: str = response.content.decode("cp932")
csv_data: str = StringIO(decoded_content)
df = pd.read_csv(csv_data).iloc[:-1, [2, 1]]
df[df.columns[0]] = df[df.columns[0]].astype(str).str[:4]

df
```
